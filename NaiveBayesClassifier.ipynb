```python
import nltk
nltk.download('all')
```

```python
from nltk.tokenize import word_tokenize
```

```python
import pandas as pd
import os
os.getcwd()
```

```python
os.chdir('/content/drive/My Drive/syntax_project')
training_source = pd.read_excel('training_source.xlsx')
```

```python
training_source_num = len(training_source)
```

```python
training_set = []

for i in range(0, training_source_num):
    training_set_1 = (training_source['초'].iloc[i], '초')
    training_set.append(training_set_1)
    training_set_2 = (training_source['중'].iloc[i], '중')
    training_set.append(training_set_2)
    training_set_3 = (training_source['고'].iloc[i], '고')
    training_set.append(training_set_3)

print(training_set)
```

```python
all_words = set(word.lower() for sentence in training_set
                   for word in word_tokenize(sentence[0]))
all_words
```

```python
all_words = []
for sentence in training_set:
    for word in word_tokenize(sentence[0]):
        all_words.append(word.lower())

set(all_words)
```

```python
t = [({word: (word in word_tokenize(x[0]))
   for word in all_words}, x[1]) for x in training_set]
t
```

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
test_sentence = 'He came not to complain, but to help us. '

test_sent_features = {word.lower() : (word in word_tokenize(test_sentence.lower()))
                                         for word in all_words}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```
