```python
from textblob import TextBlob
from textblob import Word
```

```python
import pandas as pd
import os
```

```python
# add

file_path = ''
file_name = ''

os.chdir(file_path)
poems_all = pd.read_excel(file_name)
```

```python
poems_all.head()
```

```python
poems_content = poems_all['내용']
```

```python
poems_num = len(poems_all)
```

```python
#pos_tagging
for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_each.tags)
```

```python
#sentiment_analysis
#polarity [-1.0, 1.0]
#subjectivity [0.0, 1.0](0.0 is very objective and 1.0 is very subjective)
for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_each.sentiment)
```

```python
#word_counts
for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_each.word_counts['will'])
```

```python
#pasing
for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_each.parse())
    print('\n')
```

```python
#N_Grams
for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_each.ngrams(6))
```

```python
#pos, neg classifier
from textblob.classifiers import NaiveBayesClassifier
train = [('I love this sandwich.', 'pos'), ('This is an amazing place!', 'pos'),('I feel very good about these beers.', 'pos'),('I do not like this restaurant', 'neg'),('I am tired of this stuff.', 'neg'), ("I can't deal with this", 'neg'), ("My boss is horrible.", "neg")]
cl = NaiveBayesClassifier(train)

for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_all['작품'].iloc[i])
    print(cl.classify(poems_each))
```

```python
##suffix end
for i in range(0,poems_num):
    poems_each = TextBlob(poems_all['내용'].iloc[i])
    print(poems_each.ends_with('ing'))
```
